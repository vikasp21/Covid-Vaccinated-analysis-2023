```python
plt.figure(figsize=(20,10))
plt.bar(data['Country'],data['Vaccinated'])
plt.xlabel('Country',fontsize=20)
plt.ylabel('Vaccinated',fontsize=20)
plt.title('Covid Vaccinated Report')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.plot(data['Country'],data['FullyVaccinated'])
plt.xlabel('Country')
plt.ylabel('FullyVaccinated')
plt.title('Line graph Dashboard report')
plt.grid(True)
plt.show()
```
